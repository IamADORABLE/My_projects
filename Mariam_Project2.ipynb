```python
import os
import json

# File to store tasks
TASK_FILE = "tasks.json"

# Load tasks from file
def load_tasks():
    if os.path.exists(TASK_FILE):
        with open(TASK_FILE, 'r') as f:
            return json.load(f)
    return []

# Save tasks to file
def save_tasks():
    with open(TASK_FILE, 'w') as f:
        json.dump(tasks, f, indent=4)

# Initialize tasks
tasks = load_tasks()

# Function to add a task
def add_task():
    task = input("Enter task: ")
    priority = input("Set priority (High/Medium/Low): ").capitalize()
    tasks.append({"task": task, "priority": priority, "completed": False})
    print(f'Task "{task}" added with {priority} priority!')
    save_tasks()

# Function to remove a task
def remove_task():
    view_tasks(show_completed=True)
    try:
        idx = int(input("Enter task number to remove: ")) - 1
        removed = tasks.pop(idx)
        print(f'Task "{removed["task"]}" removed!')
        save_tasks()
    except (IndexError, ValueError):
        print("Invalid task number!")

# Function to mark task as completed
def mark_completed():
    view_tasks(show_completed=False)
    try:
        idx = int(input("Enter task number to mark as completed: ")) - 1
        if not tasks[idx]["completed"]:
            tasks[idx]["completed"] = True
            print(f'Task "{tasks[idx]["task"]}" marked as completed!')
            save_tasks()
        else:
            print("Task already completed.")
    except (IndexError, ValueError):
        print("Invalid task number!")

# Function to edit a task
def edit_task():
    view_tasks(show_completed=True)
    try:
        idx = int(input("Enter task number to edit: ")) - 1
        new_task = input("Enter new task name: ")
        new_priority = input("Enter new priority (High/Medium/Low): ").capitalize()
        tasks[idx]["task"] = new_task
        tasks[idx]["priority"] = new_priority
        print("Task updated successfully!")
        save_tasks()
    except (IndexError, ValueError):
        print("Invalid task number!")

# Function to view tasks
def view_tasks(show_completed=True):
    if not tasks:
        print("No tasks to display!")
        return

    print("\n===== To-Do List =====")
    sorted_tasks = sorted(tasks, key=lambda x: {"High": 1, "Medium": 2, "Low": 3}[x["priority"]])

    for i, task in enumerate(sorted_tasks, 1):
        if show_completed or not task["completed"]:
            status = "✔" if task["completed"] else "✘"
            print(f'{i}. [{status}] {task["task"]} (Priority: {task["priority"]})')

# Main loop
while True:
    print("\n=== TO-DO LIST MENU ===")
    print("1. Add Task")
    print("2. Remove Task")
    print("3. Edit Task")
    print("4. Mark Task as Completed")
    print("5. View Tasks")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == '1':
        add_task()
    elif choice == '2':
        remove_task()
    elif choice == '3':
        edit_task()
    elif choice == '4':
        mark_completed()
    elif choice == '5':
        view_tasks()
    elif choice == '6':
        print("Goodbye! Tasks saved.")
        break
    else:
        print("Invalid choice! Please select from 1 to 6.")
```
